```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
# Loading dataset
mel_data = pd.read_csv('melb_data.csv')
mel_data.head()
```

```python
# finding missing values
missing_values = mel_data.isnull().sum()
missing_values
```

```python
mel_data.shape
```

```python
# checking total number of cells
total_cells = np.product(mel_data.shape)
total_cells
```

```python
# checking total number of missing values
total_missing_values = missing_values.sum()
total_missing_values
```

```python
# calculating the total percentage of missing value in the dataset
total_percent = (total_missing_values/ total_cells) * 100
total_percent
```

```python
y = mel_data.Price
y.head()
```

```python
mel_predictors = mel_data.drop(['Price'], axis = 1)
mel_predictors
```

```python
X = mel_predictors.select_dtypes(exclude = ['object'])
X
```

```python
# Divide into training and validation datasets
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 0, test_size = 0.2, train_size = 0.8)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_dataset(X_train, X_valid, y_train, y_valid):
  model = RandomForestRegressor(n_estimators = 10, random_state = 0)
  model.fit(X_train, y_train)
  preds = model.predict(X_valid)
  return mean_absolute_error(y_valid, preds)
```

```python
# Approach 1 (Drop columns with missing values)
cols_with_missing = [col for col in X_train.columns if X_train[col].isnull().any()]

# Drop columns in training and validation data
reduced_X_train = X_train.drop(cols_with_missing, axis = 1)
reduced_X_valid = X_valid.drop(cols_with_missing, axis = 1)

print(" MAE from Approach 1 (Drop columns with missing vallues): ")
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))
```

```python
# Approach 2 replace missing values with the mean values along with the each column
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))

# Imputation removed column names
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns

print("MAE from Approach 2 (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

```python
# Approach 3, we imputing values

X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()

for col in cols_with_missing:
  X_train_plus[col + " _was_missing"] = X_train_plus[col].isnull()
  X_valid_plus[col + " _was_missing"] = X_valid_plus[col].isnull()

my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))

# Imputation removed column names
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns

print("MAE from Approach 2 (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```
