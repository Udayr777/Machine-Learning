```python
# uploading the dataset from Google drive
mel_data_path = "/content/drive/MyDrive/Kaggle_Dataset/melb_data.csv"
```

```python
mel_data = pd.read_csv(mel_data_path)
mel_data.head()
```

```python
# Finding missing values in the dataset
missing_values = mel_data.isnull().sum()
missing_values
```

```python
mel_data.shape
```

```python
# Total number of cells in the dataset
Total_cells = np.product(mel_data.shape)
Total_cells
```

```python
# Total number of missing cells
Total_missing_cells = missing_values.sum()
Total_missing_cells
```

```python
# calculating the total missing value percent
Total_percent = (Total_missing_cells/Total_cells) * 100
Total_percent
```

```python
# Filter rows with missing price values
new_mel_data = mel_data.dropna(axis = 0)
new_mel_data
```

```python
# Choose target and features
y = new_mel_data.Price
y.head()
```

```python
mel_features = ['Rooms','Bathroom','Landsize','BuildingArea','YearBuilt','Lattitude','Longtitude']
mel_features
```

```python
X = new_mel_data[mel_features]
X
```

```python
X.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Define model
mel_model = DecisionTreeRegressor(random_state = 1)
# Fit model
mel_model.fit(X, y)
```

```python
# Calculating Mean Aboslute Error
from sklearn.metrics import mean_absolute_error
predicted_home_prices = mel_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
# split data into training and validation data, for both features and target
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state= 0)
```

```python
mel_model = DecisionTreeRegressor()
#Fit Model
mel_model.fit(train_X, train_y)
```

```python
#get predicted prices on validation data
val_predictions = mel_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
