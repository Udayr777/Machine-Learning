```python
#importing the variables
import pandas as pd
import numpy as np
```

```python
#loading the dataset of Melbourne Housing Snapshot
dataset = pd.read_csv('melb_data.csv')
dataset
```

```python
dataset.head()
```

```python
#shape of the data
dataset.shape
```

```python
#calculating the missing values in the dataset
missing_values = dataset.isnull().sum()
missing_values
```

```python
#Finding the Total Cell Values present in the dataset
Total_cell_values = np.product(dataset.shape)
Total_cell_values
```

```python
#Finding the Total Missing Values present in the dataset
Total_missing_values = missing_values.sum()
Total_missing_values
```

```python
#Finding the percentage of missing values
Total_missing_value_percentage = (Total_missing_values/Total_cell_values) * 100
Total_missing_value_percentage
```

```python
#Dropping the missing value
new_data = dataset.dropna(axis = 1)
new_data
```

```python
y = new_data.Price
y.head()
```

```python
mel_features = ['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
mel_features
```

```python
X = new_data[mel_features]
X.head()
```

```python
X.describe()
```

```python
#Bulding the model - Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
mel_model = DecisionTreeRegressor(random_state = 1)
mel_model.fit(X, y)
```

```python
#Predicting the model
print("Making predictions for the following 5 houses")
print(X.head())
print("The Predictions are")
print(mel_model.predict(X.head()))
```
